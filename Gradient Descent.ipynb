```
import numpy as np
data = np.loadtxt("CaCovidInfMarch24toMidJuly.txt")
data
```

```
#Problem 1
#By Normal Equation
def compute_cost(A, B, t, y):
    m = len(y)
    predictions = A + B*t 
    cost = (1/(2*m)) * np.sum((predictions-y)**2)
    return cost
t = np.arange(1,121)
train_t = t[:90]
test_t = t[90:]

train_data = data[:90]
test_data = data[90:]

X_train = np.vstack([np.ones(len(train_t)),train_t]).T
y_train = train_data

theta = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train

X_test = np.vstack([np.ones(len(test_t)),test_t]).T
y_test = test_data

predictions = X_test @ theta
A = theta[0]
B = theta[1]
cost = cost = compute_cost(A, B, train_t, train_data)
print(f"Optimal solution: I = {A} + {B}t")
print(f"Final Cost: {cost}")
```

```
#By full Batch Gradient Descent



A=0
B=0
alpha = 0.00001
iterations = 100000
cost_history = []
m = len(train_t)
for i in range(iterations):
    predictions = A + B*train_t 
    errors = predictions - train_data

    dA = (1/m)* np.sum(errors)
    dB = (1/m)*np.sum(errors *train_t)
    A -= alpha * dA
    B -= alpha * dB

    cost = compute_cost(A, B, train_t, train_data)
    cost_history.append(cost)

    if i %10000 == 0 :
        print(f"Iteration {i}: Cost {cost}, A {A}, B {B}")
    
print(f"Optimal solution: I = {A} + {B}t")
print(f"Final cost: {cost_history[-1]}")
```

```
#By Mini Batch 
A = 0 
B = 0
alpha = 0.00001
iterations = 100000
batch_size = 10
m = len(train_t)
cost_history = []
for i in range(iterations):
    indices = np.arange(m)
    np.random.shuffle(indices)
    train_t = train_t[indices]
    train_data = train_data[indices]

    for j in range(0,m,batch_size):
        t_batch = train_t[j:j+batch_size]
        data_batch = train_data[j:j+batch_size]

        predictions = A + B*t_batch

        errors = predictions - data_batch

        dA = (1 / batch_size) * np.sum(errors)
        dB = (1/ batch_size) * np.sum(errors* t_batch)

        A -= alpha * dA
        B -= alpha * dB

    cost = compute_cost(A, B, train_t,train_data)
    cost_history.append(cost)

    if i % 10000 == 0:
        print(f"Iteration {i}: Cost {cost}, A {A}, B {B}")

print(f"Optimal solution: I = {A} + {B}t")
print(f"Final cost: {cost_history[-1]}")
```

```
#Problem 2
#Normal Equation
def compute_cost_quad(A,B,C,t,y):
    m = len(y)
    predictions = A + B*t + C*t**2
    cost = (1/(2*m)) * np.sum((predictions-y)**2)
    return cost

X_train = np.column_stack((np.ones(train_t.shape),train_t, train_t**2))
y_train = train_data

theta = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train

X_test = np.column_stack((np.ones(test_t.shape),test_t,test_t**2))
y_test = test_data

predictions = X_test @ theta
A, B, C = theta
cost = compute_cost_quad(A, B, C, train_t, train_data)
print(f"Optimal solution: I = {A} + {B}t+{C}t²")
print(f"Final Cost: {cost}")
```

```
#By Full Batch Gradient Descent
A=0
B=0
C = 0
alpha = 0.0000001
iterations = 100000
cost_history = []
m = len(train_t)
for i in range(iterations):
    predictions = A + B*train_t + C*train_t**2
    errors = predictions - train_data

    dA = (1/m)* np.sum(errors)
    dB = (1/m)*np.sum(errors *train_t)
    dC = (1/m)*np.sum(errors * train_t**2)
    A -= alpha * dA
    B -= alpha * dB
    C -= alpha * dC

    cost = compute_cost_quad(A, B,C, train_t, train_data)
    cost_history.append(cost)

    if i %10000 == 0 :
        print(f"Iteration {i}: Cost {cost}, A {A}, B {B}, C {C}")
    
print(f"Optimal solution: I = {A} + {B}t+{C}t²")
print(f"Final cost: {cost_history[-1]}")
```

```
#By Mini Batch 
A = 0 
B = 0
C = 0
alpha = 0.0000001
iterations = 100000
batch_size = 10
m = len(train_t)
cost_history = []
for i in range(iterations):
    indices = np.arange(m)
    np.random.shuffle(indices)
    train_t = train_t[indices]
    train_data = train_data[indices]

    for j in range(0,m,batch_size):
        t_batch = train_t[j:j+batch_size]
        data_batch = train_data[j:j+batch_size]

        predictions = A + B*t_batch + C*t_batch**2

        errors = predictions - data_batch

        dA = (1 / batch_size) * np.sum(errors)
        dB = (1/ batch_size) * np.sum(errors* t_batch)
        dC = (1/batch_size) * np.sum(errors*t_batch**2)

        A -= alpha * dA
        B -= alpha * dB
        C -= alpha * dC

    cost = compute_cost_quad(A, B,C, train_t,train_data)
    cost_history.append(cost)

    if i % 10000 == 0:
        print(f"Iteration {i}: Cost {cost}, A {A}, B {B}, C {C}")

print(f"Optimal solution: I = {A} + {B}t + {C}t²")
print(f"Final cost: {cost_history[-1]}")
```

```
#Problem 3 
#Normal Equation
def compute_cost_cube(A,B,C,D,t,y):
    m = len(y)
    predictions = A + B*t + C*t**2 + D*t**3
    cost = (1/(2*m)) * np.sum((predictions-y)**2)
    return cost

X_train = np.column_stack((np.ones(train_t.shape),train_t, train_t**2,train_t**3))
y_train = train_data

theta = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train

X_test = np.column_stack((np.ones(test_t.shape),test_t,test_t**2,test_t**3))
y_test = test_data

predictions = X_test @ theta
A, B, C,D = theta
cost = compute_cost_cube(A, B, C,D, train_t, train_data)
print(f"Optimal solution: I = {A} + {B}t+{C}t²+{D}t³")
print(f"Final Cost: {cost}")
```

```
#By Full Batch Gradient Descent
A=0
B=0
C = 0
D = 0
alpha = 0.000000000001
iterations = 100000
cost_history = []
m = len(train_t)
for i in range(iterations):
    predictions = A + B*train_t + C*train_t**2 + D*train_t**3
    errors = predictions - train_data

    dA = (1/m)* np.sum(errors)
    dB = (1/m)*np.sum(errors *train_t)
    dC = (1/m)*np.sum(errors * train_t**2)
    dD = (1/m) * np.sum(errors * train_t**3)
    A -= alpha * dA
    B -= alpha * dB
    C -= alpha * dC
    D -= alpha * dD

    cost = compute_cost_cube(A, B,C, D,train_t, train_data)
    cost_history.append(cost)

    if i %10000 == 0 :
        print(f"Iteration {i}: Cost {cost}, A {A}, B {B}, C {C}, D {D}")
    
print(f"Optimal solution: I = {A} + {B}t+{C}t² + {D}t³")
print(f"Final cost: {cost_history[-1]}")
```

```
#By Mini Batch
A = 0 
B = 0
C = 0
D = 0
alpha = 0.000000001
iterations = 100000
batch_size = 10
m = len(train_t)
cost_history = []
for i in range(iterations):
    indices = np.arange(m)
    np.random.shuffle(indices)
    train_t = train_t[indices]
    train_data = train_data[indices]

    for j in range(0,m,batch_size):
        t_batch = train_t[j:j+batch_size]
        data_batch = train_data[j:j+batch_size]

        predictions = A + B*t_batch + C*t_batch**2 + D*t_batch**3

        errors = predictions - data_batch

        dA = (1 / batch_size) * np.sum(errors)
        dB = (1/ batch_size) * np.sum(errors* t_batch)
        dC = (1/batch_size) * np.sum(errors*t_batch**2)
        dD = (1/batch_size) *np.sum(errors * t_batch**2)

        A -= alpha * dA
        B -= alpha * dB
        C -= alpha * dC
        D -= alpha * dD

    cost = compute_cost_cube(A, B,C,D, train_t,train_data)
    cost_history.append(cost)

    if i % 10000 == 0:
        print(f"Iteration {i}: Cost {cost}, A {A}, B {B}, C {C}, D {D}")

print(f"Optimal solution: I = {A} + {B}t + {C}t² + {D}t³")
print(f"Final cost: {cost_history[-1]}")
```
